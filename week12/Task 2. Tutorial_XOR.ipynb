## Perceptrons as Logical Operators

```python
import pandas as pd

# Activation function (Step function)
def step_function(x):
    return int(x >= 0)

# First layer weights and biases (OR, NAND)
weight1_layer1 = [0, 0]  # Weights for OR, NAND
weight2_layer1 = [0, 0]
bias_layer1 = [0, 0]

# Second layer weights and biases (AND)
weight1_layer2 = 1
weight2_layer2 = 1
bias_layer2 = -1.5

# Inputs and correct outputs for XOR
test_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
correct_outputs = [False, True, True, False]
outputs = []

# Forward pass through the network
for test_input, correct_output in zip(test_inputs, correct_outputs):
    # First Layer Computation (OR, NAND)
    output_layer1 = [step_function(weight1_layer1[i] * test_input[0] + weight2_layer1[i] * test_input[1] + bias_layer1[i]) for i in range(2)]

    # Second Layer Computation (AND)
    linear_combination_layer2 = weight1_layer2 * output_layer1[0] + weight2_layer2 * output_layer1[1] + bias_layer2
    output_layer2 = step_function(linear_combination_layer2)

    is_correct_string = 'Yes' if output_layer2 == correct_output else 'No'
    outputs.append([test_input[0], test_input[1], output_layer1, linear_combination_layer2, output_layer2, is_correct_string])

# Print output
num_wrong = len([output[5] for output in outputs if output[5] == 'No'])
output_frame = pd.DataFrame(outputs, columns=['Input 1', 'Input 2', 'Layer 1 Output', 'Layer 2 Linear Combination', 'Final Output', 'Is Correct'])
if not num_wrong:
    print('Nice!  You got it all correct.\n')
else:
    print('You got {} wrong.  Keep trying!\n'.format(num_wrong))
print(output_frame.to_string(index=False))
```
